```python
from pathlib import Path
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image
import torch

# Setup paths
IMAGES_DIR = Path("..") / "data" / "images"

# Initialize model
processor = TrOCRProcessor.from_pretrained("eshangj/TrOCR-Sinhala-finetuned")
model = VisionEncoderDecoderModel.from_pretrained("eshangj/TrOCR-Sinhala-finetuned")

def extract_text(image_path):
    image = Image.open(image_path).convert("RGB")
    pixel_values = processor(images=image, return_tensors="pt").pixel_values
    
    with torch.no_grad():
        generated_ids = model.generate(pixel_values, max_length=128)
    
    return processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

# Process single image
image1_path = IMAGES_DIR / "image1.png"
result = extract_text(image1_path)
print(f"Extracted:  {result}")

# Process all images in folder
print("\n--- Processing all images ---")
for img_path in IMAGES_DIR. glob("*.png"):
    text = extract_text(img_path)
    print(f"{img_path.name}: {text}")
```
